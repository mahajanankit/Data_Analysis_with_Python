```python
import nltk
from nltk.corpus import nps_chat
from nltk.corpus import words
```

```python
#collection words
enwords = words.words('en')
chatwords = nps_chat.words()
```

```python
#revomving duplicates
enwords = set(enwords)
chatwords = set(chatwords)
```

```python
#Calculating unmatched words
cwords = []
for w in chatwords:
    if w.lower() in enwords:
        continue
    else:
        cwords.append(w)
```

```python
print("Total number of unusual words are: ",len(cwords))
```
