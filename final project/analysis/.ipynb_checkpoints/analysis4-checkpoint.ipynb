Importing packages

```python
import seaborn
import pandas
import matplotlib as plt
```

Setting up constant varibles

```python
rootdir = "C:/Users/Ankit Mahajan/Documents/Github/mahajan_ankit/final project/"
seaborn.set(style="ticks")
```

Retrieving data

```python
df = pandas.read_csv(rootdir+'/data/processed data/IRS_Data.csv', sep=",")
```

Grouping data for every class of income over the years

```python
df=df.groupby(["year","agi_stub"], as_index=False).sum()
```

Calculating mean income for every area

```python
df['Income']=df.a00100/ df.n1
```

Preparing data for visualization

```python
category = ["Under $10,00" ,"$10,000 under $25,000", "$25,000 under $50,000","$50,000 under $75,000",
                      "$75,000 under $100,000","$100,000 or more"]
cols = df.agi_stub.unique()
```

Setting up required parameters

```python
index = df.year.unique()
df = df.groupby(["year","agi_stub"], as_index=False).mean()
df["category"] = df["agi_stub"].replace(to_replace=cols, value=category)
```

Visulising Data

```python
g = seaborn.barplot(x="year", y="Income" , hue="category", data=df)
g.figure.suptitle("Average income of different groups.")
seaborn.plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
seaborn.plt.show()
```

Exporting data

```python
g.figure.savefig("./Analysis1 output/graph.png")
```
