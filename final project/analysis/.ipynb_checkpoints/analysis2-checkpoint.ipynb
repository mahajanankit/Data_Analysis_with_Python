# Analysis 2

Importing packages

```python
import pandas
import os
import csv
import numpy
import seaborn
import random
import collections
```

Accepting user input

```python
states = [x.strip() for x in input('states = ').split(",")]
```

Defining constants

```python
rootdir = "C:/Users/Ankit Mahajan/Documents/Github/mahajan_ankit/final project/"
```

Reading prepocessed data

```python
df = pandas.DataFrame(columns=["state","zipcode","n85300","a85300","year"])
b = pandas.DataFrame()
for subdir, dirs, files in os.walk(rootdir+"/data/raw data/IRS/"):
    for file in files:
        if file.split(".")[0] in ["2013","2014"]:
            a = pandas.read_csv(os.path.join(subdir, file), sep=",")
            a.columns = [x.lower() for x in a.columns]
            b["state"] = a["state"]
            b["zipcode"] = a["zipcode"]
            b["n85300"] = a["n85300"]
            b["a85300"] = a["a85300"]
            b["year"] = int(file.split(".")[0])
            df = pandas.concat([df,b])
df =  df[df.zipcode != 0]
```

Filtering data for selected states

```python
b = pandas.DataFrame(columns=df.columns)
for d in states:
    a = df[df['state']==d]
    b = pandas.concat([b,a])
df = b
```

Grouping data for every state, zipcode and year

```python
df = df.groupby(["state","zipcode","year"], as_index=False).sum()
```

Calculating Investment for every section

```python
df["Investment"]= df.a85300/df.n85300
```

Calculaing change in investment

```python
df["change"] = df.Investment.pct_change()
df = df.fillna(0)
df = df.replace(to_replace="inf", value=0)
```

Deleting unnecessary data

```python
df.__delitem__('n85300')
df.__delitem__('a85300')
```

Filtering data to choose the areas where there was a rise in investment

```python
df = df[df.change > 0]
```

Visulaising Data for with respect to year

```python
g = seaborn.barplot(x="state", y="Investment", hue="year", data=df)
seaborn.plt.show()
```

Saving file 

```python
g.figure.savefig("./Analysis2 output/year_wise_graph.png")
```

Graph to compare growth of Investment in different states

```python
g =seaborn.lmplot(x="Investment", y="zipcode", hue="state", data=df)

seaborn.plt.show()
```

Displaying values of chnage to support the graph

```python
s = df.groupby('state', as_index=False).mean()
s.__delitem__("zipcode")
s.__delitem__("year")
s.__delitem__("Investment")
s.sort('change',ascending=False)
```

Saving the graph to a file

```python
g.fig.savefig("./Analysis2 output/growth_graph.png")
```
