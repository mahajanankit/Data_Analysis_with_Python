Importing packages that are required

```python
import pandas
import numpy
import matplotlib.pyplot as plt
import seaborn
```

Path to the root directory of this project

```python
rootdir = "C:/Users/Ankit Mahajan/Documents/Github/mahajan_ankit/final project/"
```

Retrieving prepocessed data

```python
df = pandas.read_csv(rootdir+'/data/processed data/IRS_Data.csv', sep=",")
```

Accepting user input

```python
states = [x.strip() for x in input('states = ').split(",")]
```

Filtering data for the states that user input

```python
b = pandas.DataFrame(columns=df.columns)
for d in states:
    a = df[df['state']==d]
    b = pandas.concat([b,a])
df = b
```

Calculating the sum of each field for each state in a particular year

```python
df = df.groupby(["state","year"], as_index=False).sum()
```

Calculaing per-capital income based on taxes filed ad number of dependents they carry

```python
df["percapita_income"]=df.a00100/(df.n1+df.numdep)
df = df[["state","year","percapita_income"]]
```

Plotting a categorical graph to indicate the average rise in the percapita income of the states from 2008 to 2014

```python
g = seaborn.lmplot(x="year", y="percapita", hue="state", data=df)
g.fig.suptitle('Percapita income of the states')
seaborn.plt.show()
```

Saving output files

```python
g.savefig("./Analysis1 output/graph.png")
df.to_csv("./Analysis1 output/percapita.csv")
```
